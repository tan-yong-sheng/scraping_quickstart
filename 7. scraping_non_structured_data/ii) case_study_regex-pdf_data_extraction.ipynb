```python
import requests
import re
import fitz

# url = "https://links.sgx.com/FileOpen/1_Ascendas%20Reit%20FY2021%20Press%20Release.ashx?App=Announcement&FileID=701314"
# url = "https://repository.shareinvestor.com/rpt_view.pl/id/515563031.1/type/sgxnet"
# url = "https://repository.shareinvestor.com/rpt_view.pl/id/515563415.1/type/sgxnet" # 2H2023, don't have ;ease expiry
url = "https://repository.shareinvestor.com/rpt_view.pl/id/515564675.1/type/sgxnet"

response = requests.get(url=url, 
                        stream=True)
with fitz.open(stream=response.content, filetype="pdf") as doc:  # open document
    text = chr(12).join([page.get_text() for page in doc])
    # print(text)
    pattern1=r"(?:\dH\s|FY)((?:19|20)\d{2})"
    print(re.search(pattern=pattern1, string=text))
    print(re.search(pattern=pattern1, string=text).group(1))
    
    pattern2 = r"occupancy\s.*\s(\d{1,3}(?:\.\d+)?%)" 
    print(re.search(pattern=pattern2, string=text))
    print(re.search(pattern=pattern2, string=text).group(1))
    
    pattern3 = r"lease expiry\s.*\s(\d+(?:\.\d+)?)\syears"
    print(re.search(pattern=pattern3, string=text))
    print(re.search(pattern=pattern3, string=text).group(1))
```
