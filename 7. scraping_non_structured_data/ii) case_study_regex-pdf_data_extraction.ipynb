```python
import requests
import re
import fitz

case_study1_url = "https://repository.shareinvestor.com/rpt_view.pl/id/515491042.1/type/sgxnet" # remark: lease expiry on table
case_study1_url = "https://repository.shareinvestor.com/rpt_view.pl/id/515528669.1/type/sgxnet"
case_study2_url = "https://repository.shareinvestor.com/rpt_view.pl/id/7841407.1/type/sgxnet" # not sure if occupancy rate = 77% or 

# url = "https://links.sgx.com/FileOpen/1_Ascendas%20Reit%20FY2021%20Press%20Release.ashx?App=Announcement&FileID=701314"
# url = "https://repository.shareinvestor.com/rpt_view.pl/id/515563031.1/type/sgxnet"
# url = "https://repository.shareinvestor.com/rpt_view.pl/id/515563415.1/type/sgxnet" # 2H2023, don't have ;ease expiry
# url = "https://repository.shareinvestor.com/rpt_view.pl/id/515564675.1/type/sgxnet"
# url = "https://repository.shareinvestor.com/rpt_view.pl/id/7841407.1/type/sgxnet"

url = "https://repository.shareinvestor.com/rpt_view.pl/id/581960.1/type/sgxnet"

response = requests.get(url=url, 
                        stream=True)
with fitz.open(stream=response.content, filetype="pdf") as doc:  # open document
    text = chr(12).join([page.get_text() for page in doc])
    print(text)
    pattern1=r"(?:ended\s\d{1,2}\s\w+\s.*|\dH\s|FY)((?:19|20)\d{2})"
    matched_text1 = re.search(pattern=pattern1, string=text, flags=re.IGNORECASE)
    if matched_text1 is not None:
        print(matched_text1.group(1))
    
    pattern2 = r"portfolio[’'\s\w\n]+occupanc(?:y|ies)[\s\w\n-]+(\d{2,3}(?:\.\d+)?%)" # r"occupancy\s.*\s(\d{1,3}(?:\.\d+)?%)"
    matched_text2 = re.search(pattern=pattern2, string=text)
    if matched_text2 is not None:
        print(matched_text2.group(1))

    pattern3 = r"lease expiry[\s\w\n]+(\d+(?:\.\d+)?)\syears"
    matched_text3 = re.search(pattern=pattern3, string=text, flags=re.IGNORECASE)
    if matched_text3 is not None:
        print(matched_text3.group(1))
```
