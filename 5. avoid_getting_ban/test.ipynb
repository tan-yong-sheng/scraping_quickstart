```python
import requests
from urllib.parse import urlparse
from requests_ip_rotator import ApiGateway, EXTRA_REGIONS

from lxml import html
import hashlib
import pandas
import os
from dotenv import load_dotenv

_ = load_dotenv()


src = "https://shareinvestor.com"
login_page_url: str = "https://www.shareinvestor.com/user/login.html"
login_url_endpoint: str = "https://www.shareinvestor.com/user/do_login.html?use_https=1"
stock_quote_url: str = "https://www.shareinvestor.com/prices/counter_quote_f.html"
fundamental_page_url: str = "https://www.shareinvestor.com/fundamental/financials.html" # used by # `def get_fundamental_data`
redirect_url: list =["https://www.shareinvestor.com/sg", "https://www.shareinvestor.com/my"]
historical_prices_page_url: str = "https://www.shareinvestor.com/prices/historical_price_by_stock_f.html"
logout_url: str = "https://www.shareinvestor.com/user/do_logout.html"


headers: dict = {
          "Content-Type": "application/x-www-form-urlencoded",
          'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 \
                  (KHTML, like Gecko) Chrome/121.0.0.0 Safari/537.36'
        }

def login(session = None, username: str="", password: str="") -> None:
    # login to the shareinvestor.com website
    authenticity_token = _get_authenticity_token(session, url=login_page_url)
    # skip login into website if the user is already authenticated
    cookies = session.cookies.get_dict()
    if "si_user_profile" not in cookies:
      payload = {"utf8": "✓",
                "authenticity_token": authenticity_token,
                "name": username,
                "password": "",
                "password_m": md5_hash_encrypt(password),
                "remember": "on",
                "redirect": redirect_url[1]
                }
      login_response = session.post(url=login_url_endpoint, data=payload)
      print(login_response.url)
      login_response.raise_for_status() # raise HTTPError only if unsuccessful requests call
      # assert login_response.url in redirect_url, "You are not logged in. Please re-enter your username and password again."
    return authenticity_token

def md5_hash_encrypt(password: str="") -> str:
  # encrypt the password using MD5 algorithm
  password = password.strip().lower()
  ## Create an MD5 hash object
  md5 = hashlib.md5()
  # Update the hash object with the password
  md5.update(password.encode('utf-8'))
  ## Get the hexadecimal representation of the digest
  hashed_password = md5.hexdigest()
  return hashed_password

def _get_authenticity_token(session, url) -> str:
  # to get hidden authenticity_token (or CSRF-token)
  login_page = session.get(url)
  authenticity_token = ""
  if login_page.status_code == 200:
      ## Parse the HTML content of the page using BeautifulSoup
      tree = html.fromstring(login_page.content)
      authenticity_token = tree.xpath("//*[@id='sic_login_form_user_login']/div/input[2]/@value")[0].strip()
  return authenticity_token

def _retrieve_fundamental_page(session, security_ticker="O5RU.SI",
                                 period =5, 
                                 frequency ="fy", 
                                 ):
    fundamental_page = session.get(url=fundamental_page_url, params={"counter": security_ticker,
                                                                  "period":frequency,
                                                                  "cols": str(period+1)})
    tree = html.fromstring(fundamental_page.content)
    # Scrape website using XPath
    # Define XPath expressions for each piece of information
    reporting_period = tree.xpath('//*[@id="financials_table_pl"]/thead/tr/th[position()>=2 and position() \
                                  <= (last())]/text()[preceding-sibling::br]')
    # bug - not sure how to solve
    # timeframe = tree.xpath("//*[@id='financials_table_pl']/thead/tr/th[position()>=2 and position() \
    #                          <= (last())]/text()[following-sibling::br]")
    revenue = tree.xpath('//*[@id="financials_table_pl"]/tbody/tr[1]/td/text()') # data for TTM, and 2023 to 2019
    net_income = tree.xpath('//*[@id="financials_table_pl"]/tbody/tr[17]/td/text()') # data for TTM, and 2023 to 2019
    distributable_income = tree.xpath('//*[@id="financials_table_pl"]/tbody/tr[19]/td/text()') # data for TTM, and 2023 to 2019
    distribution_per_unit = tree.xpath('//*[@id="financials_table_pl"]/tbody/tr[25]/td/text()')  # data for TTM, 2023 to 2019
    pe_ratio = tree.xpath('//*[@id="financials_table_ratio"]/tbody/tr[13]/td/text()')  # data for TTM, 2023 to 2019
    pb_ratio = tree.xpath('//*[@id="financials_table_ratio"]/tbody/tr[17]/td/text()')
    adjusted_distribution_yield = tree.xpath('//*[@id="financials_table_ratio"]/tbody/tr[27]/td/text()')

    fundamental_data = {'reporting_period': reporting_period,
      # 'timeframe': timeframe,
      'revenue': revenue,
      'net_income': net_income,
      'distributable_income': distributable_income,
      #'Occupancy Rate': '',
      #'Lease Expiries': '',
      'distribution_per_unit': distribution_per_unit,
      'PE_ratio': pe_ratio,
      'PB_ratio': pb_ratio,
      'adjusted_distribution_yield': adjusted_distribution_yield,
      #'Current Price': ''
    }
    print(security_ticker)
    print("Before validation: ", fundamental_data)
    return fundamental_data

def close(session, authenticity_token):
    # perform logout from the shareinvestor.com website
    ## skip logout if already logout
    cookies = session.cookies.get_dict()
    if "si_user_profile" in cookies:
      payload = {"utf8": "✓",
                "authenticity_token": authenticity_token,
                }
      logout_response = session.post(url = logout_url, data=payload)
      print(logout_response.url)
      # assert logout_response.url in redirect_url, "Seems like you're not properly logout from the session. \
      #                                      Please run .close() function for your SI_WebScaper class instance."
      logout_response.raise_for_status()
```
