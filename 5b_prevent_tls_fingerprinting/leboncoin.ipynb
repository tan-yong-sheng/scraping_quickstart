## Case study: Scrape website by bypassing bot which uses TLS fingerprinting detection

### Introduction
Are you looking for a reliable way to bypass anti-bots while scraping with Python? Curl_cffi, a modified cURL version that impersonates browsers, can boost your scraper's evasion capability.


### Target website
[leboncoin.fr](https://www.leboncoin.fr) is the number one marketplace for used goods in France, which we might want to get some info from. But, this site have some heavy anti-bot protection which makes it easily trigger the anti-bot blocking when we use common scraping libraries like `requests`, `httpx` or `selenium`. Thus, in this guide, we will use `curl_cffi` which prevents **TLS and HTTP2 fingerprinting** by impersonating normal web browser configurations.

2. Github library - curl_cffi: https://github.com/lexiforest/curl_cffi

(Remark: this is a very strict website, and when you use other pure http library like `requests` or `httpx`, you could easily get blocked by some website)
