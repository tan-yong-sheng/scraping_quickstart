Objective: 
- Recognizing Occupancy rate of the REITs

```python
import requests
import re
import fitz

response = requests.get("https://repository.shareinvestor.com/rpt_view.pl/id/515488895.1/type/sgxnet", 
                        stream=True)
with fitz.open(stream=response.content, filetype="pdf") as doc:  # open document
    text = chr(12).join([page.get_text() for page in doc])
    # print(text)
    pattern = r"occupancy\s.*\s(\d{1,3}(?:\.\d+)?%)" 
    print(re.search(pattern=pattern, string=text))
    print(re.search(pattern=pattern, string=text).group(1))
```
