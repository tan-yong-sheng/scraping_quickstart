```python
import requests
import pandas
import warnings

warnings.filterwarnings("ignore")
```

```python
# keep looping to send request to mudah webpage to get JSON data
# and then combined the JSON data into data_list
idx = 0
loop = 40
data_list = []

while idx < loop:
    response = requests.get("https://search.mudah.my/v1/search", 
                            params={"category":2000, 
                                   "from":idx,
                                    "limit":40,
                                    "include": "extra_images%2Cbody", 
                                    "region":9})
    
    output = response.json()
    data = output["data"]
    loop+=40
    idx+=40
    if len(data) != 0:
        try:
            print(f"Loop count: {idx} | id: {data[0]['id']}")
            data_list.extend(data)    
        except:
            break
    else:
        break
```

```python
dfproperty = pandas.json_normalize(data_list)
dfproperty
```

```python
dfproperty.shape
```

```python
dfproperty_adj = dfproperty[["id","attributes.orig_date",
        "attributes.type_name","attributes.category_name",
        "attributes.property_type_name","attributes.region_name",
        "attributes.subarea_name","attributes.title_type_name",
        "attributes.rooms_name","attributes.bathroom_name","attributes.size",
        "attributes.price","attributes.monthly_rent"]]
dfproperty_adj.rename(columns=lambda x: x.replace("attributes.",""),
                     inplace=True)
dfproperty_adj = dfproperty_adj.drop_duplicates() # remove duplicated rows
dfproperty_adj
```

```python
dfproperty_adj.to_csv('mudah.csv',index=False)
```
