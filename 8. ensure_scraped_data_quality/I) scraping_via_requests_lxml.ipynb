#### Scrape the website content

```python
import requests
from lxml import html

def fetch_data(session, url, params, headers):
    data_list = []
    response = session.request("GET", url, params, headers)
    tree = html.fromstring(response.content)
    find_quotes_with_xpath_selector = tree.xpath("//article[@class='product_pod']")
    for find_quote in find_quotes_with_xpath_selector:
        price = find_quote.xpath("//article[@class='product_pod']//div[@class='product_price']/p[@class='price_color']/text()")
        book_title = find_quote.xpath("//article[@class='product_pod']/h3/a/text()")
        book_url = find_quote.xpath("//article[@class='product_pod']/h3/a/@href")
        star_rating = find_quote.xpath("//article[@class='product_pod']/p/@class")
        # author = find_quote.xpath("span/small[@class='author']/text()")
        # text = find_quote.xpath("span[@class='text']/text()")
        # tags = find_quote.xpath("div[@class='tags']/a/text()")
        data_list.append(
                {"price": price, 
                "book_title": book_title,
                "book_url": book_url, 
                "star_rating": star_rating
                }
            )
    return data_list
```

```python
page_num = 1
session = requests.Session()
params = {}
headers = {
  'accept': 'text/html,application/xhtml+xml,application/xml;q=0.9,image/avif,image/webp,image/apng,*/*;q=0.8,application/signed-exchange;v=b3;q=0.7',
  'user-agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/121.0.0.0 Safari/537.36'
}

# loop the page number to scrape
while True:
    try:
        url = f"https://books.toscrape.com/catalogue/page-{page_num}.html"
        output_data = fetch_data(session, url, params=params, headers=headers)
        page_num +=1
        if not output_data: # if output_data = None or False, it will stop executing the while loop
            break
        print(output_data)
    except Exception as e:
        break
```
